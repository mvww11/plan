# Regressão Linear
Métodos de regressão linear não são desenhados especificamente para lidar com forecast de séries temporais (ao contrário de modelos como os da família ARIMA, que veremos a seguir). Portanto, para usar regressão linear para fazer forecast, precisamos identificar características na série temporal e usar essas características como variáveis independentes na regressão.

Como uma primeira abordagem, vamos modelar cada cluster isoladamente. Nesse notebook estudaremos o cluster A.

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
from statsmodels.tools.eval_measures import rmse, meanabs
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm

#Importando os dados
df = pd.read_csv('dados.csv', parse_dates=['DataEntrega'], index_col='DataEntrega')

#Filtrando apenas o cluster que modelaremos nesse notebook
dependent_variable = 'Remessas_A'
volume_variable = 'Volume_A'
df_A = df[[dependent_variable, volume_variable]]
df_A.head()
```

### Seasonal Decompose
Duas características de séries temporais que podem ser usadas como variáveis independentes em modelos de regressão linear para fazer o forecast da série são o trend e a sazonalidade da série.

Existem outras features que poderíamos incorporar como variáveis independentes para modelar o cluster X, como o valor do volume do próprio cluster X, e informações sobre remessas e volume dos outros clusters. Numa primeira abordagem, vamos deixar essas variáveis de lado e trabalhar apenas com trend e sazonalidade somente do cluster que estamos modelando de forma isolada no momento.

Nossa série tem sazonalidade com período de 7 dias, conforme veremos mais detalhadamente na modelagem de time series, num outro notebook. Note também, no gráfico abaixo, que não existe uma tendência consistente de subida ou descida das remessas para o Cluster A. Isso sugere que o trend não será significante na regressão do número de remessas. Ou seja, é provável que nosso modelo de regressão linear final leve em consideração somente a sazonalidade (i.e. o dia da semana) na previsão das remessas para o Cluster A.

```python
#decompondo a série em trend, sazonalidade e resíduos:
from statsmodels.tsa.seasonal import seasonal_decompose

decomp = seasonal_decompose(df_A[dependent_variable], period=7)

#plotando
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1, figsize=(15,10))
fig.suptitle('Seasonal Decompose', fontsize=16)
ax1.set_ylabel('componente trend')
ax2.set_ylabel('componente seasonal')
ax3.set_ylabel('componente irregular')
ax4.set_ylabel('Série Original')
decomp.trend.plot(ax=ax1)
decomp.seasonal.plot(ax=ax2)
decomp.resid.plot(ax=ax3)
decomp.observed.plot(ax=ax4)
plt.show()
```

### Feature Enginnering
Vamos, agora, criar as variáveis que representam o trend e a sazonalidade da série e que serão usadas como variáveis independentes na regressão.

#### Trend
Uma abordagem clássica para criar uma variável que vai representar o trend da série é usar o a posição (ordem) da observação, i.e., essa variável assume o valor de 1 para a primeira observação (2019-10-01), 2 para a segunda observação (2019-10-02), etc. 

Fazer a regressão da série nessa variável dará uma estimativa do crescimento marginal dia a dia ao longo do intervalo estudado, já que a ordem é cronológica. Esse crescimento marginal dia a dia funciona como um proxy para o trend da série.

Criamos essa variável com o código abaixo.

```python
#criando o trend
trend = pd.DataFrame(index=df_A.index, columns=['trend'], data=np.arange(1,len(df_A.index)+1))

#concatenando o trend com nossos dados
df_A = pd.concat([df_A, trend], axis=1)
df_A
```

#### Sazonalidade
Continuando com o feature engineering, vamos criar nossa segunda variável, que representa a sazonalidade. Como nossa sazonalidade tem período semanal, nossa estratégia será criar 7 variáveis dummie, uma para cada dia da semana. Assim, a variável dummy de segunda feira vale 1 para uma observação que ocorre numa segunda feira, e 0 para uma observação que ocorre em qualquer outro dia.

Para criar essa variável, primeiro criamos uma coluna com o dia da semana (DOW) de cada observação. Em seguida, criamos as 7 variáveis dummie a partir da coluna de dia da semana:

```python
#Cria uma coluna com dia da semana:
x = df_A.reset_index()['DataEntrega'].dt.day_name().to_frame().rename(columns={'DataEntrega': 'DOW'})
x = pd.concat([df_A.reset_index()['DataEntrega'], x], axis=1)
x = x.set_index('DataEntrega', drop=True)

#dataframe com trend e dia da semana
df_A = pd.concat([df_A, x], axis=1)
df_A
```

```python
#criando as colunas dummie
dummies = pd.get_dummies(df_A.DOW)
df_A = pd.concat([df_A, dummies], axis=1).drop('DOW', axis=1)

df_A
```

A biblioteca `statsmodels` para Python (que usaremos para fazer a regressão) não tem um argumento que permita incluir o intercepto na regressão.

Assim, precisamos fazê-lo à mão. Para tal, vamos criar uma coluna que vale 1 para todas as observações. Chamamos essa coluna de `const`. Seu coeficiente será nosso intercepto na regressão.

A linguagem Python é a mais popular para fazer Data Science (~90% da comunidade de Ciência de dados usa Python, enquanto que ~10% usa R). Isso acontece porque existem excelente bibliotecas em Python para trabalhar com Machine Learning, como `scikit-learn` e `tensorflow`. Além disso, as ferramentas de cloud, em geral, são mais bem preparadas para receber modelos escritos com essas bibliotecas. O GCP, por exemplo, coloca no ar um modelo escrito em `tensorflow` com poucos cliques. Todavia, existem situações em que a linguagem R é muito mais robusta que o Python. Isso acontece, por exemplo, com regressões lineares e séries temporais. Manipular as caracterísitcas de um modelo VAR (vector autoregression) em Python é bastante difícil, enquanto que em R é relativamente fácil usando o pacote `vars`.

Como sou fluente em Python, acabo preferindo usá-lo até nessas situações, como na regressão linear que estou modelando agora. Pra quem é bom de martelo, todo problema é um prego. Se fosse em R, eu não precisaria estar incluindo o intercepto à mão. Imagino que, na Plannera, eu acabaria ganhando mais experiência com o R, uma vez que parece que é a preferência da casa e que eu modelaria séries temporais quase que semanalmente.

Com o código abaixo, adicionamos a coluna do intercepto.

```python
#adicionando uma coluna constante para o intercepto da regressão
df_A = sm.add_constant(df_A)
df_A.head()
```

## Train test split
Da mesma maneira que no modelo ARIMA, vamos separar o mês de agosto de 2020 para nosso conjunto de teste, e os meses anteriores para o conjunto de treinamento.

Assim podemos estimar a acurácia do modelo em observações nunca vistas por ele.

```python
#train test split
df_train = df_A[:'2020-07-31'].copy()
df_test = df_A['2020-08-01':].copy()
```

# Modelo 1: trend + seasonality
Em nosso primeiro modelo, utilizaremos as variáveis que representam o trend e a sazonalidade como variáveis explicativas na regressão das remessas.

```python
#pegando o trend e sazonalidade e separando as variáveis independentes (X) da dependente (y)
X_train = df_train.drop([volume_variable,dependent_variable], axis=1).copy()
y_train = df_train[dependent_variable]
X_test = df_test.drop([volume_variable,dependent_variable], axis=1)
y_test = df_test[dependent_variable]
```

```python
#treinando o modelo
model1 = sm.OLS(y_train, X_train)
model1 = model1.fit()
model1.summary()
```

#### Significância estatística
Como discutido previamente, o trend da série não é estatisticamente significativo a 5% na regressão (p-valor 0.498). Desse modo, vamos remover essa variável da regressão.

Antes disso, apenas para dar uma ideia, vamos ver qual é o MAE (mean absolute error) desse modelo no conjunto de treinamento e de teste.

```python
#Criando um dicionario para guardar os MAE de cada modelo
dict_MAE = {'MAE_training': [], 'MAE_test': []}

#fazendo as previsões do training set
predictions = model1.predict(X_train)

#criando um dataframe com previsões do training set
predictions.name = 'predictions'
df_train_predictions = pd.concat([y_train, predictions.to_frame()], axis=1)

#imprimindo o MAE no training set
print(f'MAE no training set: {meanabs(y_train, predictions)}')

#Salvando os valores de MAE desse modelo
dict_MAE['MAE_training'].append(meanabs(y_train, predictions))


################################################################################

#fazendo as previsões do test set
predictions = model1.predict(X_test)

#criando um dataframe com previsões do training set
predictions.name = 'predictions'
df_test_predictions = pd.concat([y_test, predictions.to_frame()], axis=1)

#imprimindo o MAE no training set
print(f'MAE no test set: {meanabs(y_test, predictions)}')

#Salvando os valores de MAE desse modelo
dict_MAE['MAE_test'].append(meanabs(y_test, predictions))
```

#### Acurácia
Encontramos um MAE de 113 no test set, menor que o MAE do modelo SARIMA que treinamos.

Abaixo fazemos um plot da previsão do modelo para os conjuntos de treinamento e de teste.

```python
#Gráfico das previsões
fig = plt.figure(figsize=(20, 5))
plt.plot(df_A.index, df_A[dependent_variable], label=dependent_variable)
plt.plot(df_train_predictions.predictions.index, df_train_predictions['predictions'], color='r', label='Previsão (training)')
plt.plot(df_test_predictions.predictions.index, df_test_predictions['predictions'], color='g', label='Forecast (test)')
plt.legend(loc='best', fontsize='xx-large')
plt.xticks(fontsize=18)
plt.yticks(fontsize=16)

plottitle = f'{dependent_variable} - Modelo 1'
fig.suptitle(plottitle, fontsize=25)
plt.show()
```

# Modelo 2: seasonality
Nesse segundo modelo, removemos o trend, pois determinamos acima que ele não é estatisticamente significante(a 5%) na regressão.

Em seguida, treinamos o novo modelo.

```python
#removendo o trend das nossas variáveis independentes
X_train = X_train.drop('trend', axis=1)
X_test = X_test.drop('trend', axis=1)
```

```python
#treinando o modelo 2
model2 = sm.OLS(y_train, X_train)
model2 = model2.fit()
model2.summary()
```

Todas as variáveis sazonais são estatistivamente significantes na regressão do modelo 2.

#### Interpretação do modelo 2
Como o modelo 2 tem como variáveis explicativas apenas as variáveis dummie que representam os dias da semana, isso significa que ele vai fazer a mesma previsão para todas as segundas feiras, a mesma previsão para todas as terças feiras, etc.

Apesar desse modelo, como mostrado abaixo, ter um bom MAE (97 no test set), acredito que ele não seja o modelo mais inteligente que podemos desenvolver. Mas nos dá um bom benchmark!

Abaixo fazemos o cálculo do MAE e plotamos o gráfico das previsões do modelo 2.

```python
#fazendo as previsões do training set
predictions = model2.predict(X_train)

#criando um dataframe com previsões do training set
predictions.name = 'predictions'
df_train_predictions = pd.concat([y_train, predictions.to_frame()], axis=1)


#imprimindo o MAE no training set
print(f'MAE no training set: {meanabs(y_train, predictions)}')

#Salvando os valores de MAE desse modelo
dict_MAE['MAE_training'].append(meanabs(y_train, predictions))

#fazendo as previsões do test set
predictions = model2.predict(X_test)

#criando um dataframe com previsões do training set
predictions.name = 'predictions'
df_test_predictions = pd.concat([y_test, predictions.to_frame()], axis=1)

#imprimindo o MAE no training set
print(f'MAE no test set: {meanabs(y_test, predictions)}')

#Salvando os valores de MAE desse modelo
dict_MAE['MAE_test'].append(meanabs(y_test, predictions))
```

```python
#Gráfico das previsões
inicio_intervalo = '2020-05-01'
df_plot = df_A.loc[inicio_intervalo:]

fig = plt.figure(figsize=(20, 5))
plt.plot(df_plot.index, df_plot[dependent_variable], label=dependent_variable)
plt.plot(df_train_predictions.loc[inicio_intervalo:].predictions.index, df_train_predictions['predictions'].loc[inicio_intervalo:], color='r', label='Previsão (training)')
plt.plot(df_test_predictions.predictions.index, df_test_predictions['predictions'], color='g', label='Forecast (test)')
plt.legend(loc='best', fontsize='xx-large')
plt.xticks(fontsize=18)
plt.yticks(fontsize=16)

plottitle = f'{dependent_variable} - Modelo 2'
fig.suptitle(plottitle, fontsize=25)
plt.show()
```

Perceba como temos um mesmo valor de previsão para cada dia da semana, conforme discutido acima.

# Model3: seasonality + Volume
No case, estamos supondo que a empresa já possui um modelo de previsão do volume de cada dia, com baixa margem de erro. Desse modo, é plausível criar um modelo onde o volume é variável independente na regressão das remessas.

É importante ressaltar que, mesmo que essa margem de erro seja baixa, colocar uma variável prevista como input na previsão de uma segunda variável vai gerar propagações de erro. Devemos ter cuidado, porque isso pode introduzir viés em nossa previsão (i.e. se a previsão do volume for enviesada, enviesaremos também a previsão das remessas).

Abaixo, preparamos o dataset para o modelo 3, incluindo o volume.

```python
#train test split
df_train = df_A[:'2020-07-31']
df_test = df_A['2020-08-01':]

#preparando o dataset, já sem o trend
X_train = df_train.drop([dependent_variable, 'trend'], axis=1)
y_train = df_train[dependent_variable]
X_test = df_test.drop([dependent_variable, 'trend'], axis=1)
y_test = df_test[dependent_variable]

X_test.head()
```

```python
#treinando o modelo
model3 = sm.OLS(y_train, X_train)
model3 = model3.fit()
model3.summary()
```

#### Significância estatística
O coeficiente de sexta feira não tem significância estatística (a 5%), pois encontramos um p-valor de 0.142. O procedimento estatisticamente correto é remover o dummy de sexta feira da regressão e treinar novamente o modelo. O processo de remover o coeficiente de menor significância estatística e retreinar o modelo deve ser repetido até que sobrem apenas coeficientes estatisticamente significantes.

Abaixo executamos esse procedimento:

```python
#eliminando a variável independente com menor significância estatística
X_train = X_train.drop('Friday', axis=1)
X_test = X_test.drop('Friday', axis=1)

#retreinando o modelo
model3 = sm.OLS(y_train, X_train)
model3 = model3.fit()
model3.summary()
```

Segunda feira não tem significância estatística:

```python
#eliminando a variável independente com menor significância estatística
X_train = X_train.drop('Monday', axis=1)
X_test = X_test.drop('Monday', axis=1)

#retreinando o modelo
model3 = sm.OLS(y_train, X_train)
model3 = model3.fit()
model3.summary()
```

Terça feira não tem significância estatística:

```python
#eliminando a variável independente com menor significância estatística
X_train = X_train.drop('Tuesday', axis=1)
X_test = X_test.drop('Tuesday', axis=1)

#retreinando o modelo
model3 = sm.OLS(y_train, X_train)
model3 = model3.fit()
model3.summary()
```

Quarta feira não tem significância estatística:

```python
#eliminando a variável independente com menor significância estatística
X_train = X_train.drop('Wednesday', axis=1)
X_test = X_test.drop('Wednesday', axis=1)

#retreinando o modelo
model3 = sm.OLS(y_train, X_train)
model3 = model3.fit()
model3.summary()
```

```python
#Boxplot do volume
df_A[volume_variable].plot.box()
plt.show()
```

#### Interpretação do modelo 3
Finalmente, encontramos um modelo com todos os coeficientes estatisticamente significantes (a 5%).

Interpretação do modelo:
- Temos um intercepto de 238. Isso quer dizer que nosso modelo, para prever as remessas de um dado dia, parte de um valor de base fixo e faz alterações incrementais sobre esse valor.
- Caso esse dia seja um domingo, a previsão é praticamente de zero, pois o coeficiente da dummy de domingo é praticamente igual ao intercepto em módulo, mas oposto em sinal.
- Se o dia for sábado, a previsão cai bastante em relação ao valor de base, pois o coeficiente da dummy de sábado é grande e negativo (mas não tanto quanto o de domingo).
- Se o dia for uma quinta feira, a previsão de remessas aumenta em cerca de 91 unidades. Isso aponta que nosso modelo identificou quinta feira como o dia mais movimentado.
- Nosso modelo faz um ajuste fino da previsão de remessas levando em consideração o volume do dia. Como vemos no boxplot acima, a maioria dos dias possui volume menor que 5000. Com esse volume, a correção na previsão seria de 0.0697*5000~350 unidades. Esse ajuste feito com a informação do volume permitiu, por exemplo, que o modelo 3 previsse o grande pico de remessas que aconteceu na primeira metade de junho/2020. Já o modelo 2 não conseguiu prever esse pico (comparar o gráfico de previsões do modelo anterior com o próximo gráfico).

Abaixo calculamos a acurácia do modelo.

```python
#fazendo as previsões do training set
predictions = model3.predict(X_train)

#criando um dataframe com previsões do training set
predictions.name = 'predictions'
df_train_predictions = pd.concat([y_train, predictions.to_frame()], axis=1)

#imprimindo o MAE no training set
print(f'MAE no training set: {meanabs(y_train, predictions)}')

#Salvando os valores de MAE desse modelo
dict_MAE['MAE_training'].append(meanabs(y_train, predictions))

#fazendo as previsões do test set
predictions = model3.predict(X_test)

#criando um dataframe com previsões do training set
predictions.name = 'predictions'
df_test_predictions = pd.concat([y_test, predictions.to_frame()], axis=1)

#imprimindo o MAE no training set
print(f'MAE no test set: {meanabs(y_test, predictions)}')

#Salvando os valores de MAE desse modelo
dict_MAE['MAE_test'].append(meanabs(y_test, predictions))
```

#### Avaliação do modelo
Com MAE de 77 no conjunto de teste, o modelo 3 foi o melhor modelo encontrado até agora. Todavia, como já observado, **essa acurácia está sendo subestimada**. Isso porque o algoritmo foi treinado com valores verdadeiros de volume, e em produção o input será de um valor previsto de volume. Apesar dessa previsão ter baixa margem de erro, isso certamente diminuirá a acurácia do modelo em prever as remessas, se comparada com a acurácia aqui encontrada.

Para ter uma estimativa mais realista da acurácia do modelo em produção, deveríamos ter regredido as remessas na **previsão** que do volume que estava disponível previamente. Ao utilizarmos o valor verdadeiro registrado do volume, estamos incorrendo no chamado *data leakage*: estamos usando informações no treinamento do modelo que não estarão disponíveis quando o modelo for colocado em produção. A consequência disso é que temos uma estimativa subestimada da acurácia que o modelo terá em produção.

Abaixo plotamos o gráfico das previsões do modelo 3.

```python
#Gráfico das previsões
inicio_intervalo = '2020-05-01'
df_plot = df_A.loc[inicio_intervalo:]

fig = plt.figure(figsize=(20, 5))
plt.plot(df_plot.index, df_plot[dependent_variable], label=dependent_variable)
plt.plot(df_train_predictions.loc[inicio_intervalo:].predictions.index, df_train_predictions['predictions'].loc[inicio_intervalo:], color='r', label='Previsão (training)')
plt.plot(df_test_predictions.predictions.index, df_test_predictions['predictions'], color='g', label='Forecast (test)')
plt.legend(loc='best', fontsize='xx-large')
plt.xticks(fontsize=18)
plt.yticks(fontsize=16)

plottitle = f'{dependent_variable} - Modelo 3'
fig.suptitle(plottitle, fontsize=25)
plt.show()
```

# Forecast
Faremos o forecast das remessas para o mês de setembro/2020 usando o modelo 2 (que não requer o input de volume), e usando o modelo 3 (que requer o input de volume). Não vou usar o modelo 1 porque uma das variáveis independentes não se mostrou estatisticamente significante, de modo que é possível que não represente, de fato, a estrutura da relação entre os inputs e as remessas.

Primeiramente, vamos formatar o input dos nossos modelos para o mês de setembro/2020.

```python
#criando o intervalo de tempo para o forecast
forecast_date_range = pd.date_range(start='2020-09-01', end='2020-09-30', freq='D')

#criando o dataframe com esse intervalo de tempo e uma coluna de valores 1 (coluna const)
df_forecast = pd.DataFrame(data=np.ones(len(forecast_date_range)), index=forecast_date_range, columns=['const'])
df_forecast.index.name = 'DataEntrega'

#Cria uma coluna com dia da semana:
x = df_forecast.reset_index()['DataEntrega'].dt.day_name().to_frame().rename(columns={'DataEntrega': 'DOW'})
x = pd.concat([df_forecast.reset_index()['DataEntrega'], x], axis=1)
x = x.set_index('DataEntrega', drop=True)

#adicionando a coluna de dia da semana ao df de forecast
df_forecast = pd.concat([df_forecast,x], axis=1)

#criando dummies com o dia da semana
dummies = pd.get_dummies(df_forecast.DOW)
df_forecast = pd.concat([df_forecast, dummies], axis=1).drop('DOW', axis=1)

df_forecast
```

### Forecast: Modelo 2
O dataframe acima contém o input para o modelo 2.

O forecast do modelo 2 para o mês de setembro é dado abaixo.

```python
#escolhendo o model2 para fazer o forecast
model = model2

#fazendo as previsões do mês de setembro
forecast = model.predict(df_forecast)

#criando um dataframe com o forecast
forecast.name = 'forecast'
forecast_output2 = forecast.to_frame()
forecast_output2
```

Conforme já discutido, o modelo 2 prevê sempre os mesmos valores para cada dia da semana.

### Domingos e feriados
Como sabemos que nos domingos e feriados não há envio de remessas, vamos alterar na mão a previsão para esses dias para 0. Será um modelo hídrido, portanto: parte previsto pelo modelo e parte colocado à mão. Imagino que não seja legal vender ao cliente um software que prevê que ele enviará remessas num dia em que sabe-se que nenhuma remessa será enviada.

<font color=red>Importante: o enunciado do case dizia que a operação de expedição não funcionaria aos domingos e feriados. Entretando, no forecast de Volume que a Carolina me enviou, a quantidade de remessas aos **sábados** e domingos é nula. Desse modo, o modelo 3 (que sabe que sábado tem volume = 0) fará previsões sempre menores para o sábado, se comparado com o modelo 2 (que não tem essa informação). Caso a operação de expedição realmente não funcione também aos sábados, seria interessante zerarmos o forecast de remessas para o sábado, como fizemos para domingos e feriados.</font>

```python
#Zerando a previsão de remessas em domingos e feriados
dias_off = ['2020-09-06', '2020-09-13', '2020-09-20', '2020-09-27', '2020-09-07']
for dia_off in dias_off:
    forecast_output2.loc[dia_off] = 0    
forecast_output2
```

Abaixo plotamos o forecast do modelo 2 para setembro, junto com os dados dos meses anteriores.

```python
#range do gráfico
df_plot = df_A.loc['2020-07-01':]

#Gráfico
fig = plt.figure(figsize=(20, 5))
plt.plot(df_plot[dependent_variable].index, df_plot[dependent_variable], label=dependent_variable)
plt.plot(forecast_output2.forecast.index, forecast_output2.forecast, color='r', label='Forecast')
plt.legend(loc='upper left', fontsize='xx-large')
plt.xticks(fontsize=18)
plt.yticks(fontsize=16)

plottitle = f'{dependent_variable} - Modelo 2 - Forecast'
fig.suptitle(plottitle, fontsize=25)
plt.show()
```

### Forecast: Modelo 3

Para fazer o forecast do mês de setembro usando o modelo 3, precisamos das informações do volume de cada dia durante esse mês, além dos dummies de sábado, domingo e quinta feira:
