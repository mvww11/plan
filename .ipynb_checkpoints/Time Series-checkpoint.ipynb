# Time Series
Para prever a quantidade de remessas, partiremos da família ARMA. Os modelos da família ARMA fazem a previsão da variável dependente para o tempo t olhando os valores dessa mesma variável em instantes de tempo anteriores (AR: Auto Regression), assim como os desvios dessa variável em relação a uma certa média nos intervalos de tempo anteriores (MA: Moving Averages). Historicamente, esses modelos têm se mostrado boas opções para se modelar séries temporais, com larga aceitação entre a comunidade de econometria e ciência de dados.

Para modelar nossa série, precisamos torná-la estacionária (mais sobre isso à frente). Para isso, fazemos transformações como diferenças. Nosso modelo faz a previsão, na verdade, das diferenças. A seguir, para fazer a previsão da própria variável de interesse, fazemos a soma (integração) das diferenças, partindo do valor inicial. Para automarizar o processo de diferenciação e integração, usamos o conveniente modelo ARIMA.

A série de remessas é sazonal, como veremos adiante. Por isso, vamos utilizar uma adaptação do modelo ARIMA para incorporar sazonalidade, conhecida como SARIMA (Seasonal ARIMA). Ademais, para levar em consideração o volume expedido como variável exógena na previsão do número de remessas, usamos o modelo SARIMAX (o "X" vem de eXogenous), que facilita essa incorporação.

Como uma primeira abordagem, vamos modelar cada cluster isoladamente. Nesse notebook estudaremos o cluster A.

Abaixo, importamos as bibliotecas e nosso dados.

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
%matplotlib inline

#Importando os dados
df = pd.read_csv('dados.csv', parse_dates=['DataEntrega'], index_col='DataEntrega')

#Filtrando apenas cluster que modelaremos nesse notebook
dependent_variable = 'Remessas_A'
volume_variable = 'Volume_A'
df_A = df[[dependent_variable, volume_variable]]
df_A.head()
```

## Características da série de remessas
Plotando a série de remessas abaixo, vemos que ela tem uma característica sazonal marcante. Imagino que o período de sazonalidade seja de 7 dias, pois faz sentido para um serviço de expedição de remessas.

Ao mesmo tempo, a série parece não apresentar uma tendência consistente de crescimento ou decrescimento.

```python
#Gráfico variável dependente
inicio_intervalo = '2020-05-01'
df_plot = df_A.copy()

fig = plt.figure(figsize=(20, 5))
plt.plot(df_plot.index, df_plot[dependent_variable], label=dependent_variable)
plt.xticks(fontsize=18)
plt.yticks(fontsize=16)

plottitle = f'{dependent_variable}'
fig.suptitle(plottitle, fontsize=25)
plt.show()
```

## Estacionariedade
Os modelos ARIMA pressupõem estacionariedade. Isso porque, se a série não é estacionária, parâmetros como média e variância tem diferentes valores a cada intervalo de tempo. Desse modo, teríamos um número de parâmetros a encontrar sempre maior do que o número de observações à disposição. Com mais parâmetros para ajustar do que observações disponíveis, a modelagem não poderia ser feita com modelos de regressão como os ARIMA.

Vimos no plot acima, por inspeção visual, que nossa série apresenta forte comportamento sazonal. Uma série sazonal não pode ser estacionária, por definição, uma vez que parâmetros como sua média são função do valor de tempo, e, portanto, não constantes.

Para determinar qual é o período de sazonalidade da série, vamos observar o gráfico do ACF (auto correlation function). O ACF mede a correlação do valor da variável no tempo t com o valor dessa própria variável shiftada de x dias (ou lags). Se existe forte auto correlação com um certo lag x, isso significa que os valores da série tendem se repetir a cada x dias.

```python
#plotando o ACF:
fig, ax1 = plt.subplots()
plot_acf(df_A[dependent_variable], lags=25, zero=False,  ax=ax1)
plt.show()
```

### Período da sazonalidade
Pelo acf acima, confirmamos nossa intuição inicial de que a sazonalidade é semanal (observe como os picos do acf estão em lags múltimos de 7).

## Seasonal decompose
De posse do período da sazonalidade, fazemos abaixo a decomposição da série entre componente de trend, componente de sazonalidade e componente irregular.

Observe como o trend não possui comportamento consistente cresce ou decrescente, conforme notamos no plot inicial da série.

```python
#decompondo a série em trend, sazonalidade e resíduos:

decomp = seasonal_decompose(df_A[dependent_variable], period=7)

#plotando
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1, figsize=(15,10))
fig.suptitle('Seasonal Decompose', fontsize=16)
ax1.set_ylabel('componente trend')
ax2.set_ylabel('componente seasonal')
ax3.set_ylabel('componente irregular')
ax4.set_ylabel('Série Original')
decomp.trend.plot(ax=ax1)
decomp.seasonal.plot(ax=ax2)
decomp.resid.plot(ax=ax3)
decomp.observed.plot(ax=ax4)
plt.show()
```

## Teste ADF
o teste ADF (augmented dickey-fuller) é uma ferramenta para verificar se a série é estacionária. O teste avalia a hipótese nula de que a série é um random walk (portanto não estacionária), contra a hipótese alternativa de que a série é estacionária.

Fazemos esse teste abaixo. Obtivemos um p-valor de 0.0013. Assim, curiosamente, o teste ADF aponta que a série é estacionária (a 5% de significância). Como é possível que o teste diga que a série é estacionária, se vimos que ela é sazonal e , portanto, não estacionária? Isso acontece porque o teste ADF testa a estacionariedade de trend, apenas. Esse teste não consegue capturar uma não estacionariedade devido à sazonalidade, como é nosso caso.

Portanto, faremos as diferenças com a intenção de remover a sazonalidade da série e, assim, torná-la estacionária. Faremos as diferenças com lag de 7 dias, uma vez que determinamos que o período da sazonalidade é de 7 dias. Em outras palavras, pegaremos cada valor da nossa série temporal no tempo t e subtrairemos o valor da série no tempo (t-7).

```python
#teste ADF
result = adfuller(df_A[dependent_variable])
#printando o p-valor:
print('p-valor: ', result[1])
```

```python
#tirando as diferenças com lag de 7 meses e jogando fora os valores nulos gerados
rio_df_diff = df_A[dependent_variable].copy()
rio_df_diff = df_A[dependent_variable].diff(7)
rio_df_diff = rio_df_diff.dropna()

#novo teste ADF
result = adfuller(rio_df_diff)
#printando o p-valor:
print('p-valor: ', result[1])

#plotando a série das diferenças de 12 meses
rio_df_diff.plot(figsize=(14,5))
plt.show()
```

### Série estacionária!
Após fazermos as diferenças com lag de 7 dias e plotar o resultado acima, parece que a série ficou estacionária! Por inspeção visual, não parece que há sazonalidade na série das diferenças. Isso é confirmado por um novo teste ADF, também acima, que dá p-valor de e-11!

Agora, seguros de que nossa série está estacionária, podemos partir para a modelagem.

## Train Test split
Vamos separar o mês de agosto de 2020 para nosso conjunto de teste, e os meses anteriores para o conjunto de treinamento.

Assim podemos estimar a acurácia do modelo em observações nunca vistas por ele.

```python
#train test split
df_train = df_A[:'2020-07-31']
df_test = df_A['2020-08-01':]
```

# Modelo 1: SARIMA
O primeiro modelo que vamos treinar é um Seasonal ARIMA, ou SARIMA.
Esse modelo tem dois termos: o primeiro é idêntico ao ARIMA, e o segundo é o termo sazonal. Esse termo sazonal é muito parecido com o ARIMA, pois modela o valor de y no tempo t com base nos valores de y em tempos anteriores e no valor do desvio em relação a um valor esperado nos tempos anteriores. A diferença é que, ao invés de usar o n lags anteriores, o modelo era usar apenas lags múltiplos da sazonalidade, que aqui é 7.

Portanto, nosso modelo SARIMA tem 6 ordens: (p,d,q) e (P,D,Q), além do período da sazonalidade, que aqui é 7.
- d é quantas diferenciações simples (lag 1) fizemos para tornar a série estacionária. No nosso caso é zero.
- D é quantas diferenciações com lag igual ao período da sazonalidade fizemos para tornar a série estacionária. No nosso caso é 1.
- p e q são as ordens do AR e MA, respectivamente, no termo não sazonal.
- P e Q são as ordens do AR e MA, respectivamente, no termo sazonal.

Estamos com 4 ordens ainda a escolher: p,q,P,Q. Assim, vamos treinar diferentes modelos com valores candidatos a cada um desses hiperparâmetros. Ao final, utilizaremos o critério de informação de Akaike (AIC) para determinar qual é o melhor modelo.

Para fazer essa busca de forma automarizada, utilizaremos a biblioteca `pmdarima`.

```python
import pmdarima as pm

# treinando diferentes modelos e guardando o de menor AIC
model1 = pm.auto_arima(df_train[dependent_variable],
                      seasonal=True, m=7,
                      d=0, D=1,
                      start_p=0, start_q=0, start_P=0, start_Q=0,
                      max_p=1, max_q=0, max_P=6, max_Q=0,
                 	  max_order=None,
                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True)
print(model1.summary())                    
```

obs: o modelo não passa no teste de Ljung-Box. Pode ser que exista correlação entre os resíduos, violando um pressuposto do modelo. Analisarei isso.

### Escolhendo o melhor modelo
O AIC mede um trade-off entre complexidade e acurácia do modelo. Quanto menor o AIC, melhor é esse trade-off.

O modelo que possui menor AIC, entre os experimentados, é o SARIMA(1,0,0)(6,1,0)7. Assim, o termo não sazonal é AR com p=1, e o termo sazonal é AR com P=6. q e Q valem, ambos, zero.

#### Interpretação do modelo
A interpretação de nosso modelo é que, para prever o valor de y no tempo t, esse modelo leva em consideração o valor de y nos tempos (t-1), (t-7), (t-14), (t-21), (t-28), (t-35) e (t-42).

Podemos perceber que, de fato, a sazonalidade é de extrema importância para o modelo. Para fazer a previsão das remessas numa certa quinta feira, por exemplo, o modelo precisa olhar apenas para o dia anterior e para as 6 quintas feiras anteriores.

A seguir, vamos treinar o melhor modelo usando a biblioteca `statsmodels`, para depois fazer as previsões no conjunto de teste.

```python
#treinando o melhor modelo
from statsmodels.tsa.statespace.sarimax import SARIMAX
model1 = SARIMAX(df_train[dependent_variable], order=(1,0,0), trend='c', seasonal_order=(6,1,0,7))
model1 = model1.fit()
```

```python
print(model1.summary())
```

### Acurácia do modelo
A seguir, fazemos a previsão do modelo para o conjunto de teste e medimos sua acurácia pelo MAE (Mean absolute error).

Decidi usar o MAE, e não o RMSE (Root mean squared error), porque esse último dá uma punição maior ao modelo por errar outliers. Avaliando a série, existem picos que são absolutamente imprevisíveis dado o conjunto de dados que temos à disposição (e.g. o pico em 2020-06). Não desejo punir o modelo de forma mais intensa por esse tipo de erro, portanto usarei o MAE.

Encontramos um MAE de 147 no conjunto de treinamento e de 128 no conjunto de teste.

Como é possível que o desempenho no conjunto de treinamento seja pior que no conjunto de teste? Isso acontece porque, no conjunto de treinamento, a série apresenta de forma mais frequente um comportamento irregular que não pode ser previsto pelo modelo (e.g. o pico em 2020-06). Portanto, é esperado que o MAE no conjunto de treinamento seja consistentemente maior que o MAE no conjunto de teste entre todos os modelos desenvolvidos no case.

Um gráfico com as previsões no conjunto de treinamento (in sample) e no conjunto de teste (out of sample) é mostrado abaixo.

```python
#MAE no conjunto de treinamento
MAE = np.mean(np.abs(model1.resid))
print(f"Mean Absolute Error no conjunto de treinamento: {MAE}\n")

#Fazendo a previsão no conjunto de teste
predictions_test = model1.predict(start=(len(df_train[dependent_variable])), end=(len(df_train[dependent_variable])+len(df_test[dependent_variable])-1))

#MAE no conjunto de teste
MAE_test = np.mean(np.abs(predictions_test - df_test[dependent_variable]))
print(f"Mean Absolute Error no conjunto de teste: {MAE_test}\n")
```

```python
#df de previsões in sample (train)
predictions_train = model1.predict(start=0, end=(len(df_train[dependent_variable])-1))
predictions_train.name = 'predictions_train'
predictions_train = predictions_train.to_frame()

#df de previsões out of samples (test)
predictions_test = model1.predict(start=(len(df_train[dependent_variable])), end=(len(df_train[dependent_variable])+len(df_test[dependent_variable])-1))
predictions_test.name = 'predictions_test'
predictions_test = predictions_test.to_frame()

#Gráfico das previsões
fig = plt.figure(figsize=(20, 5))
plt.plot(df_A.index, df_A[dependent_variable], label=dependent_variable)
plt.plot(predictions_train.index, predictions_train['predictions_train'], color='r', label='Previsão (training)')
plt.plot(predictions_test.index, predictions_test['predictions_test'], color='g', label='Previsão (test)')
plt.legend(loc='best', fontsize='xx-large')
plt.xticks(fontsize=18)
plt.yticks(fontsize=16)

plottitle = f'{dependent_variable} - Modelo 1 - SARIMA'
fig.suptitle(plottitle, fontsize=25)
plt.show()
```

# Modelo 2: SARIMAX
Podemos estender o modelo acima para incorporar uma variável exógena, como o volume. Assim, nosso modelo vai prever o valor do número de remessas no tempo t com base em valores anteriores do número de remessas e com base também no valor do volume no tempo t.

<font color=red>Importante: No notebook onde fizemos o modelo de regressão linear, já discutimos qual é o impacto que a incorporação do volume tem na estimativa da acurácia do modelo de previsão de remessas. Como o volume também está sendo previsto, ainda que com baixa margem de erro, a estimativa que teremos da acurácia do modelo SARIMAX de previsão de remessas utilizando nosso conjunto de teste estará substimada.</font>

Abaixo, usamos novamente o pacote `pmdarima` para varrer diferentes ordens de modelo e encontrar aquela com menor AIC.

```python
exogenous = df_train[volume_variable].to_frame()

# treinando diferentes modelos e guardando o de menor AIC
model2 = pm.auto_arima(df_train[dependent_variable], exogenous,
                      seasonal=True, m=7,
                      d=0, D=1,
                      start_p=0, start_q=0, start_P=0, start_Q=0,
                      max_p=1, max_q=0, max_P=6, max_Q=0,
                 	  max_order=None,
                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True)
print(model2.summary())                    
```

#### Interpretação do modelo
Nosso SARIMAX (com volume como input para a previsão das remessas) ótimo possui as seguintes ordens: SARIMAX(0,0,0)(6,1,0)7.

A interpretação de nosso modelo é que, para prever o valor de y no tempo y, esse modelo leva em consideração o valor de y nos tempos (t-7), (t-14), (t-21), (t-28), (t-35) e (t-42), **além do volume no tempo t**.

Vamos treinar o SARIMAX usando a biblioteca `statsmodels` para, em seguida, fazer as previsões em nosso conjunto de treinamento e de teste.

```python
#from statsmodels.tsa.statespace.sarimax import SARIMAX
model2 = SARIMAX(df_train[dependent_variable], df_train[volume_variable],order=(1,0,0), trend='c', seasonal_order=(6,1,0,7))
model2 = model2.fit()
```

```python
print(model2.summary())
```

### Acurácia do modelo
Abaixo, encontramos um MAE de 107 no conjunto de treinamento e de 105 no conjunto de teste. Em comparação com o modelo modelo 1, isso sugere que o modelo 2, que incorpora o volume, é mais acurado. Entretanto, confirme discutido, **a acurácia do modelo 2 no conjunto de teste subestima a acurácia que teremos com esse modelo em produção**.

Um gráfico com as previsões no conjunto de treinamento (in sample) e no conjunto de teste (out of sample) é mostrado mais abaixo.

```python
#MAE no conjunto de treinamento
MAE = np.mean(np.abs(model2.resid))
print(f"Mean Absolute Error no conjunto de treinamento: {MAE}\n")

#Fazendo a previsão no conjunto de teste
predictions_test = model2.predict(start=(len(df_train[dependent_variable])), end=(len(df_train[dependent_variable])+len(df_test[dependent_variable])-1), exog=df_test[volume_variable])

#MAE no conjunto de teste
MAE_test = np.mean(np.abs(predictions_test - df_test[dependent_variable]))
print(f"Mean Absolute Error no conjunto de teste: {MAE_test}\n")
```

```python
#df de previsões in sample (train)
predictions_train = model2.predict(start=0, end=(len(df_train[dependent_variable])-1), exog=df_train[volume_variable])
predictions_train.name = 'predictions_train'
predictions_train = predictions_train.to_frame()

#df de previsões out of samples (test)
predictions_test = model2.predict(start=(len(df_train[dependent_variable])), end=(len(df_train[dependent_variable])+len(df_test[dependent_variable])-1), exog=df_test[volume_variable])
predictions_test.name = 'predictions_test'
predictions_test = predictions_test.to_frame()

#Gráfico das previsões
fig = plt.figure(figsize=(20, 5))
plt.plot(df_A.index, df_A[dependent_variable], label=dependent_variable)
plt.plot(predictions_train.index, predictions_train['predictions_train'], color='r', label='Previsão (training)')
plt.plot(predictions_test.index, predictions_test['predictions_test'], color='g', label='Previsão (test)')
plt.legend(loc='best', fontsize='xx-large')
plt.xticks(fontsize=18)
plt.yticks(fontsize=16)

plottitle = f'{dependent_variable} - Modelo 1 - SARIMA'
fig.suptitle(plottitle, fontsize=25)
plt.show()
```

# Forecast para setembro/2020
Abaixo está o forecast das remessas para o mês de setembro usando o modelo 1, SARIMA, (que não requer o input de volume), e usando o modelo 2, SARIMAX, (que requer o input de volume).

```python
#Importando a previsão de volume para o mês de setembro
df_volume = pd.read_csv('dados_volume_setembro2020.csv', parse_dates=['DataEntrega'], index_col='DataEntrega')
#juntando os volumes registrados em agosto com os volumes previstos para setembro (necessário como input do modelo 2)
exogenous_forecast = pd.concat([df_A['2020-08-01':][volume_variable], df_volume[volume_variable]]).to_frame()

#df de forecast model 1
predictions_forecast_1 = model1.predict(start=(len(df_train[dependent_variable])+len(df_test[dependent_variable])), end=(len(df_train[dependent_variable])+len(df_test[dependent_variable]))+30-1)
predictions_forecast_1.name = 'model1_forecast'
predictions_forecast_1 = predictions_forecast_1.to_frame()
predictions_forecast_1

#df de forecast model 2
predictions_forecast_2 = model2.predict(start=(len(df_train[dependent_variable])+len(df_test[dependent_variable])), end=(len(df_train[dependent_variable])+len(df_test[dependent_variable]))+30-1, exog=exogenous_forecast)
predictions_forecast_2.name = 'model2_forecast'
predictions_forecast_2 = predictions_forecast_2.to_frame()
predictions_forecast_2

#imprimindo os forecasts
df_forecast = pd.concat([predictions_forecast_1, predictions_forecast_2], axis=1)
df_forecast
```

#### Domingos e feriados
Assim como fizemos na modelagem de Regressão Linear, vamos zerar os forecasts de remessas para domingos e feriados. Veja no notebook da regressão linear a discussão que fizemos sobre sábado.

```python
#Zerando a previsão de remessas em domingos e feriados
dias_off = ['2020-09-06', '2020-09-13', '2020-09-20', '2020-09-27', '2020-09-07']
for dia_off in dias_off:
    df_forecast['model1_forecast'].loc[dia_off] = 0    
    df_forecast['model2_forecast'].loc[dia_off] = 0    
df_forecast
```

Por fim, plotamos abaixo o forecast de cada modelo, para comparação.

```python
df_plot = df_A['2020-07-01':]
#Gráfico das previsões
fig = plt.figure(figsize=(20, 5))
plt.plot(df_plot.index, df_plot[dependent_variable], label=dependent_variable)
plt.plot(df_forecast.index, df_forecast['model1_forecast'], color='r', label='model 1 (SARIMA)')
plt.plot(df_forecast.index, df_forecast['model2_forecast'], color='g', label='model 2 (SARIMAX)')
plt.legend(loc='upper left', fontsize='xx-large')
plt.xticks(fontsize=18)
plt.yticks(fontsize=16)

plottitle = f'{dependent_variable} - Forecast - Time Series'
fig.suptitle(plottitle, fontsize=25)
plt.show()
```
